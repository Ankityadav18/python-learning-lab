```python
#1-D dictionary
d={"name":"ankit","age":24,"gender":"male"}
print(d)
```

```python
#2-D dictionary
d={
  "name":"ankit",
  "age":24,
  "gender":"male",
  "sem":4,
  "subject":{
    "dsa":90,
    "python":96
  }

  }

print(d)
```

```python
print(dict)
```

```python
#usinf dict() function
d4=dict(name="ankit",age=24,gender="male")
print(d4)
```

```python
#using duplicate keys in dictionary
d5={"name":"ankit","name":"amit"}
print(d5) #it will print the last key value pair
```

```python
#immutable item as key in dictionary
d6={"name":"ankit",(1,2,3):23}
print(d6)
```

```python
#accessing items from dictionary
print(d["name"])
```

```python
print(d4[0]) #it will give error because we cannot access dictionary items by index
```

```python
# adding new key value pair in dictionary
d4["city"]="delhi"
print(d4)
```

```python
# nesting adding new key value pair in dictionary
d["subject"]["java"]=90
print(d)
```

```python
#Removing key value pair from dictionary
#pop() method
d4.pop("city")
print(d4)
```

```python
#item() method
d4.popitem()
print(d4)
```

```python
# deleting key value pair from dictionary
del d4["age"]
print(d4)
```

```python
d12={
  "name":"ankit",
  "age":24,
  "gender":"male",
  "sem":4,
  "subject":{
    "dsa":90,
    "python":96
  }

  }
print(d12["subject"]["dsa"]) #accessing nested dictionary value
```

```python
#Membership operators in dictionary
print("name" in d12) #True
```

```python
#iterating through dictionary
for key in d12:
  print(key) #it will print all the keys in dictionary
```

```python
#Dictionary functions
#len() function
print(len(d12)) #it will print the length of dictionary
```

```python
#sorted() function
print(sorted(d12)) #it will print the sorted keys of dictionary
```

```python
#min() function
print(min(d12)) #it will print the minimum key in dictionary
```

```python
#max() function
print(max(d12)) #it will print the maximum key in dictionary
```

```python
#items() function
print(d12.items()) #it will print all the key value pairs in dictionary
```

```python
#values() function
print(d12.values()) #it will print all the values in dictionary
```

```python
#update() function
d12.update({"name":"amit"})
print(d12) #it will update the value of key "name" to "amit"
```

```python
#dictionary comprehension
#syntax: {key:value for (key,value) in iterable}

#print the square of all the numbers in a list using dictionary comprehension
numbers=[1,2,3,4,5]
squared={num:num**2 for num in numbers}
print(squared)
```
