```python
d12={
  "name":"ankit",
  "age":24,
  "gender":"male",
  "sem":4,
  "subject":{
    "dsa":90,
    "python":96
  }

  }
print(d12["subject"]["dsa"]) #accessing nested dictionary value
```

```python
#Membership operators in dictionary
print("name" in d12) #True
```

```python
#iterating through dictionary
for key in d12:
  print(key) #it will print all the keys in dictionary
```

```python
#Dictionary functions
#len() function
print(len(d12)) #it will print the length of dictionary
```

```python
#sorted() function
print(sorted(d12)) #it will print the sorted keys of dictionary
```

```python
#min() function
print(min(d12)) #it will print the minimum key in dictionary
```

```python
#max() function
print(max(d12)) #it will print the maximum key in dictionary
```

```python
#items() function
print(d12.items()) #it will print all the key value pairs in dictionary
```

```python
#values() function
print(d12.values()) #it will print all the values in dictionary
```

```python
#update() function
d12.update({"name":"amit"})
print(d12) #it will update the value of key "name" to "amit"
```

```python
#dictionary comprehension
#syntax: {key:value for (key,value) in iterable}

#print the square of all the numbers in a list using dictionary comprehension
numbers=[1,2,3,4,5]
squared={num:num**2 for num in numbers}
print(squared)
```

```python
#zip 
keys=["name","age"]
values=["ankit",24]
result=zip(keys,values)
print(result) #it will print the zip object
```

```python
print(dict(result)) #it will print the dictionary created from zip object
```

```python
number1=[1,2,3]
cube={num:num**3 for num in number1}
print(cube)
```

```python
names=["ankit","amit","mohit"]
lenth={name:len(name) for name in names}
print(lenth)
```

```python
OddEven={num:("Even" if num%2==0 else "odd") for num in range(1,11)}
print(OddEven)
```

```python
Even={num:num**2 for num in range(1,21) if num%2==0}
print(Even)
```

```python
Odd={num:num**3 for num in range(1,21) if num%2!=0}
print(Odd)
```

```python
names = ["ankit", "rahul", "aman"]
ages = [24, 25, 23]
result ={name:age for name,age in zip(names,ages)}
print(result)
```
