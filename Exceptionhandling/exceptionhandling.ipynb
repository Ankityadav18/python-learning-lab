```python
class Bank:
  def __init__(self,balance):
    self.balance=balance
  def withdraw(self,amount):
    if amount<0:
      raise Exception("Amount cannot be negative")
    if self.balance<amount:
      raise Exception ("MONEY IS NOT AVAILABLE")  
    self.balance=self.balance-amount
obj=Bank(10000)
try:
  obj.withdraw(-20)
except Exception as e:
  print(e) 
else:
  print(obj.balance)  

```

```python
class MyException(Exception):
  def __init__(self,messgae):
    print(messgae)
class Bank:
  def __init__(self,balance):
    self.balance=balance
  def withdraw(self,amount):
    if amount<0:
      raise MyException("Amount cannot be negative")
    if self.balance<amount:
      raise MyException ("MONEY IS NOT AVAILABLE")  
    self.balance=self.balance-amount
obj=Bank(10000)
try:
  obj.withdraw(8000)
except MyException as e:
  print(e) 
else:
  print(obj.balance)
```
