```python
#Reading files
# 1.using read()

f=open("sample1.txt","r")
s=f.read()
print(s)
f.close()
```

```python
# how to read only particular characters
f=open("sample1.txt",'r')
s=f.read(10)
print(s)
f.close()
```

```python
#read line by line
f=open("sample1.txt","r")
print(f.readline())
print(f.readline())
```

```python
# if we have to cut down the spaces
f=open("sample1.txt","r")
print(f.readline(),end="")
print(f.readline(),end="")
print(f.readline(),end="")
```

```python
#when to use which? read for smaller and readfile for bigger file
# example where i will give about readline

f=open("sample1.txt","r")
while True:
   data=f.readline()
   if data=="":
      break
   else:
      print(data,end="")
f.close()      
```

```python
#using Context Manager(with)
#with keyword closes the file as soon as the usage is over
with open("sample.txt","w") as f:
  f.write("Ankit codes")


```

```python
#moving within file -> first 10 character then 10 characters
with open("sample1.txt","r") as f:
  print(f.read(10))
  print(f.read(10))
```

```python
#seek and tell function
with open("sample1.txt","r") as f:
  print(f.read(10))
```

```python
with open("sample1.txt","r") as f:
  print(f.read(10))
  print(f.tell())
```
