```python
#when to use which? read for smaller and readfile for bigger file
# example where i will give about readline

f=open("sample1.txt","r")
while True:
   data=f.readline()
   if data=="":
      break
   else:
      print(data,end="")
f.close()      
```

```python
#serialization:python data-> JSON
#deserialization : JSON-> Python data
```

```python
#serialization using json module
#list
import json
L=[1,2,3,4,5]
with open('demo.json','w') as f:
  json.dump(L,f)
```

```python
#dict
import json
d={
  "name":"Ankit",
  "age":24,
  "gender":"male"
}
with open("demo1.json","w") as f:
  json.dump(d,f)
```

```python
#deserialization
import json
with open('demo.json','r') as f:
  d=json.load(f)
  print(d)
  print(type(d))
```
