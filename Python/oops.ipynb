```python
#without class and objects
def display(name,age,gender):
    print("Name:", name)
    print("Age:", age)
    print("gender:", gender) 
def study(name):
  print(name, "is studying")
def play(name):
  print(name, "is playing")  

display("Ankit",20,"Male")
study("Ankit")
play("Ankit")

#Everytime we want to create a new student we have to write the same code again and again which is not a good practice. So, to overcome this problem we can use class and objects

display("Rooh",21,"Female")
study("Rooh")
play("Rooh")

display("Heer",22,"Female")
study("Heer")
play("Heer")

# here you can see that we are always passing the name,age,gender,but if we use class and objects we can create a student class and then create objects of that class for each student. This way we can avoid writing the same code again and again.
```

```python
#with class and objects
class Students:
    #constructor
    def __init__(self,name,age,gender):
        self.name = name
        self.age = age
        self.gender = gender

    def display(self):
        print("Name:", self.name)
        print("Age:", self.age)
        print("Gender:",self.gender)

    def study(self):
        print(self.name, "is studying")    


    def play(self):
        print(self.name, "is playing")
s=Students("Amit",22,"Male")
s.display()
s.study()
s.play()
# here you have seen that on calling the methods we are not always passing the name of the student as an argument. This is because we have already stored the name of the student in the object s. So, we can access it directly using self.name,self.age and self.gender.This is the main advantage of using classes and objects. We can store the data in the object and access it directly without passing it as an argument every time we call a method.


s1=Students("Raaha",24,"Female")
s1.display()
s1.study()
s1.play()

#here you can see that we have created two objects s and s1 of the class Students. Each object has its own data and methods. We can create as many objects as we want and each object will have its own data and methods. This is the main advantage of using classes and objects. We can create multiple objects of the same class and each object will have its own data and methods.
#morover,we do not have to pass the same data in the methods again and again. We can store the data in the object and access it directly using self.name,self.age and self.gender.
```

```python
class Employee:
  def __init__(self,name,id,salary):
    self.nm=name
    self.id=id
    self.sal=salary
  def display(self,new_name):
    self.nm=new_name
    print(f"The name of the employee is {new_name} and his id is {self.id} and the salary is {self.sal}")
e1=Employee("Shalu",26,95000) 

e1.nm="Ankit"

print(e1.__dict__)

e1.display("Amit")
print(e1.__dict__)
```

```python
#non-parameterized constructor
class Bank:
  def __init__(self):
    self.salary=20000
    self.name="ankit"
  def display(self):
    print(f"the name is {self.name} and the salary is {self.salary}")
b=Bank() 
print(b.__dict__)
b.display() 
```

```python
#parameterized constructor
class Banker:
  def __init__(self,name,salary):
    self.nm=name
    self.sal=salary
  def display(self):
    print(f"the name is {self.nm} and the salary is {self.sal}")
br=Banker("Ankit",100000) 
print(br.__dict__)
br.display()
```

```python
#builtin class function

class House:
  def __init__(self,room,kitchen,hall):
    self.rm=room
    self.kt=kitchen
    self.hl=hall

  def display(self):
    print(f"the house has {self.rm}Rooms,{self.kt}Kitchens and {self.hl}hall")
h=House(8,2,4)   
print(h.__dict__)
h.display()
print(getattr(h,"rm"))
(setattr(h,"rm",10))
print(h.__dict__)
delattr(h,"kt")
print(h.__dict__)
print(hasattr(h,"rm"))
```
